```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import csv
import html5lib
from pandas.io.html import read_html
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)
df = df[0]
df.head()
```

```python
df.drop(df[df.Borough == 'Not assigned'].index,inplace=True)
df.head()
```

```python
df.replace("Not assigned", df['Borough'], inplace=True)
print(df.shape)
df.head()
```

```python
grouped = df.groupby(['Postcode','Borough']).agg(lambda x: ',' .join(x)).reset_index()
print(grouped.shape)
print(grouped.columns)
df.head()
```
